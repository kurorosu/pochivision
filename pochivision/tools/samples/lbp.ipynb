```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage.feature import local_binary_pattern
from skimage.transform import resize
```

```python
def extract_lbp_features(image, P=8, R=1, method='uniform', resize_shape=(128, 128)):
    """
    LBPを使った特徴量抽出（機械学習用）
    - 回転不変LBP
    - 固定サイズにリサイズ（スケール統一）
    - 正規化ヒストグラム出力
    """
    if image.ndim == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    image = resize(image, resize_shape, anti_aliasing=True)

    lbp = local_binary_pattern(image, P, R, method=method)
    n_bins = int(lbp.max() + 1)  # 実際のラベル数で固定次元化

    hist, _ = np.histogram(lbp.ravel(), bins=n_bins, range=(0, n_bins), density=True)
    return hist
```

```python
image_path = "sample.bmp"
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

features = extract_lbp_features(image)
print("説明変数の次元:", len(features))

plt.bar(range(len(features)), features, align='center')
plt.xlabel('LBP Bin')
plt.ylabel('Normalized Frequency')
plt.title('LBP Feature Histogram')
plt.show()
```
